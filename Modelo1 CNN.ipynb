```python
'''Libreria para preprocesar las imagenes y poner el tama√±o de las imagenes'''
from tensorflow.keras.preprocessing.image import ImageDataGenerator

'''Creacion de las propiedades del entrenamiento'''
train_datagen = ImageDataGenerator(rescale=1 / 255.0,
                                   rotation_range=20,
                                   zoom_range=0.05,
                                   width_shift_range=0.05,
                                   height_shift_range=0.05,
                                   shear_range=0.05,
                                   horizontal_flip=True,
                                   fill_mode="nearest",
                                   validation_split=0.20)
'''Creacion de las propiedades de la prueba de entrenamiento'''
test_datagen = ImageDataGenerator(rescale=1 / 255.0,
                                  validation_split=0.20)
```

```python
'''Libreria para importar archivo zip'''
import zipfile
'''Se carga el dataset en forma de archivo .zip '''
with zipfile.ZipFile('dataset.zip', 'r') as zip_ref:
    zip_ref.extractall('.')
```

```python
'''Mostrar la ruta de las imagenes''' 
import os
for dirname, _, filenames in os.walk('.'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
'''Cargar el dataset de entrenamiento'''
train_dir = "./dataset/training/"
'''Cargar el dataset de prueba'''
test_df = "./dataset/test/" 
'''Se asigna tama√±o de alto'''
img_height = 250
'''Se asigna tama√±o ancho'''
img_width = 250
batch_size = 32
'''Mostrar las clases'''
os.listdir(train_dir)
```

```python
'''Se crea e'''
train_generator = train_datagen.flow_from_directory(
    directory=train_dir,
    class_mode='categorical',
    color_mode='rgb',
    subset="training",
    seed=123,
    target_size=(img_height, img_width),
    batch_size=batch_size)

valid_generator = test_datagen.flow_from_directory(
    directory=test_df,
    class_mode='categorical',
    color_mode='rgb',
    subset="validation",
    seed=123,
    target_size=(img_height, img_width),
    batch_size=batch_size)
```

```python
'''Importar tensorflow'''
import tensorflow as tf

'''Creacion del modelo'''
model = tf.keras.Sequential([
    tf.keras.applications.MobileNetV2(include_top=False,
                                      input_shape=(250, 250, 3),
                                      weights="imagenet",
                                      pooling='avg',
                                      classes=1000),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(5, activation='softmax')
])

'''Resumen del modelo'''
model.summary()
```

```python
'''Compilacion del modelo con los hiperparametros'''
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
'''Finalmente se entrena esta red neurona en 5 capas'''
history = model.fit(train_generator, validation_data=valid_generator, epochs=5)
```
